```
from pyspark.sql.functions import col
import re

def sanitize_column_names(df):
    for col_name in df.columns:
        new_name = re.sub('[^a-zA-Z0-9_]', '_', col_name.strip())
        df = df.withColumnRenamed(col_name, new_name)
    return df

# Leitura dos CSVs
covid_worldwide_df = spark.read.option("header", True).option("inferSchema", True).csv("/FileStore/covid_worldwide.csv")
vaccinations_df = spark.read.option("header", True).option("inferSchema", True).csv("/FileStore/country_vaccinations_by_manufacturer.csv")

# Renomeando colunas para evitar erros
covid_worldwide_df = sanitize_column_names(covid_worldwide_df)
vaccinations_df = sanitize_column_names(vaccinations_df)

# Criando tabelas no schema bronze
covid_worldwide_df.write.mode("overwrite").saveAsTable("bronze.covid_worldwide")
vaccinations_df.write.mode("overwrite").saveAsTable("bronze.vaccinations_by_manufacturer")
```

```
from pyspark.sql import functions as F

# Lê os dados da silver
df_vacinas = spark.table("silver.vaccinations_by_manufacturer")

# Faz a agregação com cast e renomeia as colunas
df_gold = df_vacinas.groupBy("location", "vaccine") \
    .agg(F.sum(F.col("total_vaccinations").cast("bigint")).alias("total_vacinas")) \
    .withColumnRenamed("location", "pais") \
    .withColumnRenamed("vaccine", "fabricante")

# Grava no gold
df_gold.write.mode("overwrite").saveAsTable("gold.vacinas_total_por_pais_fabricante")

```

```
from pyspark.sql.functions import col

# Carregar a silver
df = spark.table("silver.covid_worldwide")

# Conversão de colunas para tipo numérico
df_gold = df.select(
    col("Country").alias("pais"),
    col("Total_Cases").cast("bigint").alias("total_casos"),
    col("Total_Deaths").cast("bigint").alias("total_mortes"),
    col("Population").cast("bigint").alias("population")
)

# Filtrar valores nulos
df_gold = df_gold.filter(
    col("total_casos").isNotNull() & 
    col("total_mortes").isNotNull() & 
    col("population").isNotNull()
)

# Criar database gold se não existir
spark.sql("CREATE DATABASE IF NOT EXISTS gold")

# Salvar como tabela gold
df_gold.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("gold.casos_mortes_por_pais")

```

```
from pyspark.sql.functions import col

# Carregar as tabelas Gold
df_casos = spark.table("gold.casos_mortes_por_pais")
df_vacinas = spark.table("gold.vacinas_total_por_pais_fabricante")

# Realizar o join com base no país
df_flat = df_vacinas.join(
    df_casos,
    df_vacinas["pais"] == df_casos["pais"],
    "inner"
).select(
    df_vacinas["pais"],
    df_vacinas["fabricante"],
    df_vacinas["total_vacinas"],
    df_casos["total_casos"],
    df_casos["total_mortes"],
    df_casos["population"].alias("populacao")
)

# Criar database se não existir
spark.sql("CREATE DATABASE IF NOT EXISTS gold")

# Salvar como tabela Gold final (flat)
df_flat.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("gold.vacinas_casos_mortes_flat")
```

```
df_casos = spark.table("gold.casos_mortes_por_pais")

df_casos.orderBy(col("total_casos").desc()).select("pais", "total_casos", "total_mortes").show(10)

```

```
df_vacinas = spark.table("gold.vacinas_total_por_pais_fabricante")

df_vacinas.groupBy("pais") \
    .agg(_sum("total_vacinas").alias("total_vacinas")) \
    .orderBy(col("total_vacinas").desc()) \
    .show(10)
```

```
df_total_vacinas = df_vacinas.groupBy("pais") \
    .agg(_sum("total_vacinas").alias("total_vacinas"))

df_taxa = df_total_vacinas.join(df_casos, on="pais", how="inner") \
    .withColumn("taxa_vac_por_habitante", (col("total_vacinas") / col("population"))) \
    .select("pais", "total_vacinas", "population", "taxa_vac_por_habitante") \
    .orderBy(col("taxa_vac_por_habitante").desc())

df_taxa.show(10)
```

```
df_criticos = df_total_vacinas.join(df_casos, on="pais", how="inner") \
    .withColumn("taxa_mortalidade", col("total_mortes") / col("total_casos")) \
    .withColumn("taxa_vac_por_habitante", col("total_vacinas") / col("population")) \
    .filter((col("taxa_mortalidade") > 0.05) & (col("taxa_vac_por_habitante") < 0.5)) \
    .orderBy(col("taxa_mortalidade").desc())

df_criticos.show(10)
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

janela = Window.partitionBy("pais").orderBy(col("total_vacinas").desc())

df_fab_mais_usado = df_vacinas.withColumn("rank", row_number().over(janela)) \
    .filter(col("rank") == 1) \
    .select("pais", "fabricante", "total_vacinas")

df_fab_mais_usado.show(10)
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar a tabela como DataFrame
df = spark.table("gold.vacinas_casos_mortes_flat")

# Agrupar e somar vacinas por país
top_vacinas = df.groupBy("pais") \
    .agg({"total_vacinas": "sum"}) \
    .orderBy("sum(total_vacinas)", ascending=False) \
    .limit(10)

# Converter para Pandas para visualização
top_vacinas_pd = top_vacinas.toPandas()

# Plot
plt.figure(figsize=(12,6))
sns.barplot(data=top_vacinas_pd, x="pais", y="sum(total_vacinas)", palette="Blues_d")
plt.title("Top 10 Países com Mais Vacinas Aplicadas")
plt.xlabel("País")
plt.ylabel("Total de Vacinas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
